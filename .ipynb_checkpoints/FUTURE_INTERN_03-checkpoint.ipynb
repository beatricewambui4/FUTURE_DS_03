```python
#checking age distribution to know the best bands to have for the age column
df['Age'].describe()
```

```python
#creating age bands
df['Age_Bands']=pd.cut(
    df['Age'],
    bins=[0,25,40,60,100],
    labels=['Youth','Young_Adult','Middle_Age','Senior_Age']
)
```

```python
df['Education'].unique()
```

```python
#renaming the column default to credit default
df.rename(columns={'Default':'Credit_Default'},inplace=True)
```

```python
#checking call distribution to understand how will group or creat bins for the column
df['Duration'].describe()
```

```python
df['Call_Duration_Bands']=pd.cut(
    df['Duration'],
    bins=[0, 100, 300, 1000, 5000],
    labels=['Short Call',
        'Medium Call',
        'Long Call',
        'Very Long Call']
)
```

```python
df['Campaign'].unique()
```

```python
#checking campaign distibution
df['Campaign'].describe()
```

```python
#creating campaign number of calls distribution
df['Call_Frequency_Band']=pd.cut(
    df['Campaign'],
    bins=[0,1,3,10,60],
    labels=['Low Call Frequency',
        'Moderate Call Frequency',
        'High Call Frequency',
        'Very High Call Frequency'
           ]
)
```

```python
df['Previous'].unique()
```

```python
#creating previous bands
df['No_of_previous_contacts']=pd.cut( 
    df['Previous'],
    bins=[-1,0,2,5,df['Previous'].max()],
    labels=['No Contact','Low Engagement','Moderate Engagement','High Engagement']
)
```

```python
df['Pdays'].unique()
```

```python
#creating pdays bands
# first create the default category
df['pdays_band'] = 'Never Contacted'

# apply bands for customers who were contacted before
mask = df['Pdays'] != 999

df.loc[mask, 'pdays_band'] = pd.cut(
    df.loc[mask, 'Pdays'],
    bins=[0, 7, 30, df['Pdays'].max()],
    labels=[
        'Recent Contact',
        'Moderate',
        'Long Ago'
    ],
    include_lowest=True
)
```

```python
df.head()
```

```python
#analysing the data 
df['Age_Bands'].value_counts()
```

```python
df['Y'].value_counts()
```

```python
#which age group are converting the most
df.groupby('Age_Bands')['Y'].value_counts()
conversion_rate = pd.crosstab(
    df['Age_Bands'],
    df['Y'],
    normalize='index'
) * 100

conversion_rate
```

```python
#profession distribution
df['Job'].value_counts()
```

```python
#Which professions are easiest to convert?
df.groupby('Job')['Y'].value_counts()
conversion_rate = pd.crosstab(
    df['Job'],
    df['Y'],
    normalize='index'
) * 100

conversion_rate
```

```python
#Marital status distribution
df['Marital'].value_counts()
```

```python
#between married people and single which ones are most likely to convert 
df.groupby('Marital')['Y'].value_counts()
conversion_rate = pd.crosstab(
    df['Marital'],
    df['Y'],
    normalize='index'
) * 100

conversion_rate
```
